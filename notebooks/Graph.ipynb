```python
import matplotlib.pyplot as plt
import numpy as np

# Nombres de los modelos
modelos = [
    "RoBERTa", "Fine Tunning OpenAI 4o-mini", "Fine Tunning Gemini 2.0 Flash"
]

# Métricas correspondientes
precision = [0.8462, 0.8962, 0.9180]
recall    = [0.8569, 0.8712, 0.9308]
f1_score  = [0.8514, 0.8832, 0.9242]
accuracy  = [0.9506, 0.9645, 0.9530]

# Configurar posición de barras
x = np.arange(len(modelos))
width = 0.2

# Crear la figura
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x - 1.5*width, precision, width, label='Precision', color='#227c9d')
ax.bar(x - 0.5*width, recall,    width, label='Recall',    color='#ffcb77')
ax.bar(x + 0.5*width, f1_score,  width, label='F1-Score',  color='#48cfae')
ax.bar(x + 1.5*width, accuracy,  width, label='Accuracy',  color='#fe6d73')

# Etiquetas y diseño con texto más grande
ax.set_ylabel('Score', fontsize=14)
ax.set_title('Comparación de modelos por métricas', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(modelos, rotation=0, ha='center', fontsize=12)
ax.set_yticks(np.arange(0.5, 1.05, 0.05))
ax.tick_params(axis='y', labelsize=12)
ax.set_ylim(0.5, 1.0)
ax.legend(loc='lower right', fontsize=12)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
labels = ['No crítico (0)', 'Crítico (1)']
sizes = [50, 50]
colors = ['#66b3ff', '#ff6666']
explode = (0.01, 0.01)

# Crear la gráfica
plt.figure(figsize=(6, 5))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    startangle=90,
    textprops={'fontsize': 12}  # Tamaño del texto dentro de la gráfica
)
plt.axis('equal')
plt.title('Distribución de Hallazgos Críticos', fontsize=14)  # Tamaño del título
plt.tight_layout()
plt.show()
```
